# 1. COUNT PLOT

```python
# sns.countplot(df, x=df['SEX'])

sns.countplot(df, x=df['SEX'])
plt.xticks(rotation=90)
plt.show()
```

```python
gender_count = df.groupby('SEX').agg(count=('SEX', 'size'))     
print(gender_count) 
```

```python
sns.countplot(df, x=df['HOSP_TYPE'])
plt.xticks(rotation=90)
plt.show()
```
